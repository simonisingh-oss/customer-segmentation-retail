**Customer Tier Generation & AI Recommendation System**

---

**The objective of this notebook is to:

Convert numerical clusters into business-friendly customer segments.

Analyze each customer segment.

Generate personalized marketing recommendations.

Prepare the final dataset for the Streamlit dashboard.**

---

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Customer_Segments.csv")
```

```python
df["Cluster"].value_counts().sort_index()
```

```python
df.head()
```

**Cluster Distribution**


---

```python
cluster_count = df["Cluster"].value_counts().sort_index()

print(cluster_count)
```

```python
cluster_summary = (
    df.groupby("Cluster")
      .mean()
      .round(2)
)

cluster_summary
```

**Assign Business-Friendly Segment Names**

From your cluster statistics, the clusters can reasonably be interpreted as:

Cluster	Interpretation	Segment
0	High Income + High Spending -	VIP Customers

1	High Income + Low Spending	- Potential Customers

2	Medium Income + Medium Spending	- Regular Customers

3	Young + Good Spending	- Young Shoppers

4	Low Income + Low Spending	- Budget Customers

```python
segment_map = {
    0: "VIP Customers",
    1: "Potential Customers",
    2: "Regular Customers",
    3: "Young Shoppers",
    4: "Budget Customers"
}

df["Segment"] = df["Cluster"].map(segment_map)
```

```python
recommendation_map = {

    "VIP Customers":
    "Offer premium membership, exclusive rewards and early access to new collections.",

    "Potential Customers":
    "Provide personalized product recommendations and loyalty offers to increase spending.",

    "Regular Customers":
    "Encourage repeat purchases through reward points and bundle discounts.",

    "Young Shoppers":
    "Promote trending products, seasonal collections and student offers.",

    "Budget Customers":
    "Provide discount coupons, cashback offers and festival sales."
}

df["Recommendation"] = df["Segment"].map(recommendation_map)
```

```python
df.head()
```

```python
segment_summary = (
    df.groupby("Segment")
      .agg(
          Customers=("Cluster","count"),
          Average_Age=("Age","mean"),
          Average_Income=("Annual Income (k$)","mean"),
          Average_Spending=("Spending Score (1-100)","mean")
      )
      .round(2)
)

segment_summary
```

**10. Business Insights (Markdown)**

**Observation 1**

VIP Customers have relatively higher income and spending behaviour. They should receive premium services to improve customer retention.

**Observation 2**

Potential Customers have high purchasing power but lower spending. Personalized marketing campaigns can encourage higher purchases.

**Observation 3**

Young Shoppers show good spending behaviour despite relatively lower income. Trend-based marketing is likely to be effective.

**Observation 4**

Budget Customers spend less and have lower income. Discounts and cashback offers can help increase their engagement.

**Observation 5**

Regular Customers form a stable customer base and should be retained through loyalty programs.
