#** EXPLORATORY DATA ANALYSIS**

1. IMPORT LIBRARIES



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. LOAD DATASET



```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

#       1.       UNIVARIATE ANALYSIS



### Age Distribution



```python
plt.figure(figsize=(7,5))
sns.histplot(df["Age"], bins=15, kde=True, color="lavender")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

Observation:
Most customers are between 20 and 40 years of age.

### Annual Income Distribution

```python
plt.figure(figsize=(7,5))
sns.histplot(df["Annual Income (k$)"], bins=15, kde=True, color="pink")
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Number of Customers")
plt.show()
```

Observation:
Customers have different income levels, with most falling in the middle-income range.






### Spending Score Distribution

```python
plt.figure(figsize=(7,5))
sns.histplot(df["Spending Score (1-100)"], bins=15, kde=True, color="orange")
plt.title("Spending Score Distribution")
plt.xlabel("Spending Score")
plt.ylabel("Number of Customers")
plt.show()
```

Observation:
Customers have a wide range of spending scores, indicating different shopping habits.

### Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

Observation:
The dataset contains both male and female customers in nearly equal numbers.

# 2.          BIVARIATE ANALYSIS

### Income vs Spending Score

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    data=df,
    color="red"
)

plt.title("Income vs Spending Score")
plt.show()
```

Observation:
Customers with similar income levels often have different spending scores.

### AGE VS SPENDING SCORE

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="Age",
    y="Spending Score (1-100)",
    data=df,
    color="purple"
)

plt.title("Age vs Spending Score")
plt.show()
```

Observation:
Younger customers generally have higher spending scores than older customers.

### Gender vs Spending Score

```python
plt.figure(figsize=(6,5))
sns.boxplot(
    x="Gender",
    y="Spending Score (1-100)",
    data=df
)

plt.title("Gender vs Spending Score")
plt.show()
```

Observation:
Both male and female customers show a wide variation in spending scores.



# 3. CORRELATION ANALYSIS

numeric-only dataset

```python
numeric_df = df.select_dtypes(include=np.number)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

Observation:
No strong correlation exists among the numerical variables in the dataset.
