```python
import streamlit as st
```

```python
df = pd.read_csv("Customer_Segments.csv")
```

**Page Configuration**

```python
st.set_page_config(
    page_title="Retail Customer Segmentation",
    layout="wide",
    page_icon="🛍️"
)
```

**Title**

```python
st.title("🛍️ Retail Customer Segmentation & Recommendation System")

st.write("Machine Learning Based Customer Segmentation")
```

**Sidebar**

```python
menu = st.sidebar.selectbox(
    "Navigation",
    [
        "Dashboard",
        "Dataset",
        "Customer Segments",
        "Recommendations",
        "Customer Search",
        "About"
    ]
)
```

**Dashboard**

```python
if menu=="Dashboard":

    st.subheader("Project Overview")

    c1,c2,c3,c4 = st.columns(4)

    c1.metric("Customers",len(df))

    c3.metric("Clusters",df["Cluster"].nunique())
    c4.metric("Features",len(df.columns))
```

**Customer Segments**

**Recommendations**

**Customer Search**

```python
menu = st.sidebar.selectbox(
    "Navigation",
    ["Dashboard", "Dataset", "Customer Segments", "Recommendations", "Customer Search", "About"]
)

if menu == "Dashboard":
    st.header("Dashboard")

    st.write(df.head())

elif menu == "Dataset":

    st.header("Dataset")

    st.dataframe(df)

    st.subheader("Statistics")

    st.write(df.describe())

elif menu == "Customer Segments":

    st.header("Customer Segments")

elif menu == "Recommendations":

    st.header("Recommendations")

elif menu == "Customer Search":

    st.header("Customer Search")

elif menu == "About":

    st.header("About")
```
