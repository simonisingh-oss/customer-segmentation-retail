# **BUISNESS UNDERSTANDING AND LITERATURE REVIEW**

### **Smart Retail Customer Segmentation and Personalized Marketing Recommendation System**

# **PROBLEM STATEMENT**

Retail businesses collect a large amount of customer data every day through purchases and transactions. However, simply collecting data is not enough. Many businesses find it difficult to analyze this information and understand the different needs and preferences of their customers. As a result, they often use the same marketing strategy for everyone, which may not attract or retain customers effectively.

Customer segmentation helps solve this problem by grouping customers based on similar characteristics such as age, annual income, and spending behavior. In this project, the Mall Customers dataset is analyzed using machine learning techniques to identify different customer groups. These segments can help businesses design personalized marketing strategies, improve customer satisfaction, and make better business decisions.





# **OBJECTIVES**

The main objectives of this project are:

* To analyze customer data and understand their purchasing behavior.


* To identify different customer segments based on age, annual income, and spending score.


* To perform customer segmentation using the K-Means Clustering algorithm.


* To help businesses understand their customers more effectively.


* To support personalized marketing by identifying customers with similar characteristics.


* To provide useful insights that can improve customer satisfaction and business growth.

# **BUISNESS UNDERSTANDING**

### 1. **What is Retail Analytics?**

Retail analytics is the process of collecting and analyzing customer and sales data to understand shopping patterns and improve business decisions. It helps retailers identify customer preferences, manage inventory, increase sales, and provide better services. By studying customer behavior, businesses can create effective marketing strategies and improve customer satisfaction.

### **2.  Why Customer Segmentation?**

Customer segmentation means dividing customers into different groups based on their age, income, and spending behavior. Every customer has different needs and shopping habits, so treating all customers in the same way is not effective. By identifying similar customer groups, businesses can provide personalized offers, improve customer experience, and increase customer loyalty.

### **3. buisness challenges**

Retail businesses face several challenges while understanding customer behavior. Large amounts of customer data make manual analysis difficult and time-consuming. It is also challenging to identify valuable customers and design personalized marketing strategies. Without proper customer segmentation, businesses may lose potential customers and miss opportunities to improve sales and customer satisfaction.

# **LITERATURE REVIEW**

Several researchers have applied machine learning techniques to improve customer segmentation and retail analytics. Türkmen (2022) compared different clustering algorithms and found that K-Means produced meaningful customer segments for marketing applications. John et al. (2023) further compared multiple clustering methods and reported that the Gaussian Mixture Model (GMM) achieved better segmentation performance than K-Means on the Online Retail dataset.

Recent studies have also focused on combining customer segmentation with sales prediction and business intelligence. Lewaaelhamd et al. (2024) demonstrated that data transformation techniques improve clustering quality, while Alharbi et al. (2025) combined customer segmentation with sales prediction models to support business decisions. Azharani and Nuryana (2026) integrated Business Intelligence tools with K-Means clustering to improve sales analysis and customer segmentation. These studies show that machine learning techniques can effectively support customer analysis, but there is still scope to improve personalization, real-time analytics, and the use of diverse customer information.

# **PROPOSED SOLUTION**

Based on the identified research gaps, this project proposes a customer segmentation system using the K-Means Clustering algorithm on the Mall Customers dataset.
