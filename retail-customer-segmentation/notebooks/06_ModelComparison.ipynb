# **MODEL COMPARISON**

##**import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

#**load dataset**

```python
X = pd.read_csv("ML_Ready_Dataset.csv")
X
```

#**Train the K-Means Model**

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=42
)

cluster_labels = kmeans.fit_predict(X)
```

##**Add Cluster Labels**

```python
customer_segments = X.copy()

customer_segments["Cluster"] = cluster_labels

customer_segments.head()
```

##**Number of Customers in Each Cluster**

```python
cluster_count = customer_segments["Cluster"].value_counts().sort_index()

print("Customers in Each Cluster")
print(cluster_count)
```

##**Cluster Centers**

```python
cluster_centers = pd.DataFrame(
    kmeans.cluster_centers_,
    columns=X.columns
)

print("Cluster Centers")
cluster_centers
```

##**Plot Cluster Distribution**

```python
plt.figure(figsize=(8,5))

cluster_count.plot(
    kind="bar"
)

plt.title("Number of Customers in Each Cluster")
plt.xlabel("Cluster")
plt.ylabel("Number of Customers")

plt.show()
```

```python
customer_segments.to_csv(
    "Customer_Segmentation_Result.csv",
    index=False
)

print("Customer segmentation results saved successfully.")
```
